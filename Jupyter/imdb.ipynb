```python
#import dataset
df_imdb = pd.read_csv(r'Desktop\File\DQLab\final projects\Dataset\IMDB\The 500 Greatest Films of All Time - IMDb.csv')
pd.set_option('display.max_columns', None)
df_imdb.head(20)
```

data that we want to get : **title, release year, type, duration, rating, total user rating, genre, director, writers, stars, achievements, details, box office**

```python
#slicing dataset
imdb = df_imdb.loc[:,'title':'achievements']
imdb
```

Getting data details and box office, note : because we see randomized shuffled data (between text 17, details and box office) columns, we have to clean it

```python
#cleaning columns Text17,details,box_office
df_clean = df_imdb.copy()
df_clean = df_clean.loc[:,['title','Text16','Text17','details','box_office']]
df_clean
```

```python
df_clean = df_clean.drop_duplicates(subset='title',keep='first')
```

```python
df_clean.info()
```

```python
df_clean['details_0'] = df_clean['Text16'].apply(lambda x : x if 'Details' in str(x) else '')
```

```python
df_clean['details_1'] = df_clean['Text17'].apply(lambda x : x if 'Details' in str(x) else '')
```

```python
df_clean['details_2'] = df_clean['details'].apply(lambda x : x if 'Details' in str(x) else '')
```

```python
df_clean['details_3'] = df_clean['box_office'].apply(lambda x : x if 'Details' in str(x) else '')
```

```python
df_clean['details_movie'] = df_clean['details_0']+df_clean['details_1']+df_clean['details_2']+df_clean['details_3']
```

```python
df_clean
```

Lakukan hal yang sama pada box office untuk mendapat info budget dan revenue movies

```python
df_clean['box_office_0'] = df_clean['Text16'].apply(lambda x : x if 'Box' in str(x) else '')
df_clean['box_office_1'] = df_clean['Text17'].apply(lambda x : x if 'Box' in str(x) else '')
df_clean['box_office_2'] = df_clean['details'].apply(lambda x : x if 'Box' in str(x) else '')
df_clean['box_office_3'] = df_clean['box_office'].apply(lambda x : x if 'Box' in str(x) else '')
df_clean
```

```python
df_clean['info_financial'] = df_clean['box_office_0']+df_clean['box_office_1']+df_clean['box_office_2']+df_clean['box_office_3']
df_clean
```

```python
#slicing with only take title,details_movie and info_financial
df_clean = df_clean.loc[:,['title','details_movie','info_financial']]
df_clean
```

```python
df_clean['languages'] = df_clean['details_movie'].str.extract('Language.*\n(.*)')
df_clean
```

```python
df_clean['first_language'] = df_clean['languages'].str.extract(r'([A-Z][a-z]*)')
```

```python
df_clean
```

```python
df_clean[df_clean['languages'].isnull()]
```

```python
#assumption for language NaN is English language
df_clean['first_language'].fillna('English (predicted)', inplace=True)
df_clean
```

```python
df_clean['countries'] = df_clean['details_movie'].str.extract('Countr.*\n(.*)')
df_clean
```

```python
df_clean['country'] = df_clean['countries'].str.extract(r'([A-Z][a-z]* [A-Z][a-z]*|[A-Z][a-z]*)')
df_clean
```

```python
df_clean['country'].fillna('Unfetch Data',inplace=True)
```

```python
df_clean
```

```python
df_clean['budget']=df_clean['info_financial'].str.extract(r'Budget\$(.*)\s\(estimated\)')
df_clean
```

```python
df_clean['gross_worldwide']=df_clean['info_financial'].str.extract(r'Gross worldwide.([0-9,]*)')
df_clean
```

```python
df_clean = df_clean.dropna(subset='budget')
```

```python
df_clean = df_clean.dropna(subset='gross_worldwide')
```

```python
df_clean
```

```python
df_clean = df_clean.loc[:,['title','first_language','countries','country','budget','gross_worldwide']]
```

```python
df_clean
```

```python
imdb_movies = pd.merge(imdb,df_clean,left_on='title',right_on='title',how='inner')
imdb_movies
```

```python
imdb_movies = imdb_movies.drop_duplicates(subset='title')
```

```python
imdb_movies = imdb_movies.loc[:,['title','release_year','type','duration','rating','total_user_rating','genre','director','writers_1','writers_2','stars_1','stars_2','stars_3','achievements','first_language','country','budget','gross_worldwide']]
```

```python
imdb_movies
```

```python
imdb_movies.info()
```

```python
imdb_movies[imdb_movies['title'].isnull()]
```

```python
#remove data that hasn't name
imdb_movies = imdb_movies.dropna(subset='title')
```

```python
imdb_movies
```

```python
imdb_movies.info()
```

```python
imdb_movies = imdb_movies.drop_duplicates(keep='first')
imdb_movies
```

```python
imdb_movies['title']=imdb_movies['title'].str.lower()
imdb_movies
```

```python
imdb_movies['release_year']=imdb_movies['release_year'].astype(int)
imdb_movies
```

```python
imdb_movies.info()
```

```python
imdb_movies['duration (hour to min)']=imdb_movies['duration'].str.extract(r'(\d)h')
imdb_movies['duration (hour to min)']=imdb_movies['duration (hour to min)'].fillna(0)
imdb_movies['duration (hour to min)']=imdb_movies['duration (hour to min)'].astype(int) * 60
imdb_movies['duration (minutes)']=imdb_movies['duration'].str.extract(r'\dh (\d*)m')
imdb_movies['duration (minutes)']=imdb_movies['duration (minutes)'].fillna(0)
imdb_movies['duration (minutes)']=imdb_movies['duration (minutes)'].astype(int)
imdb_movies['duration']=imdb_movies['duration (hour to min)']+imdb_movies['duration (minutes)']
imdb_movies
```

```python
imdb_movies['total_user_rating_0'] = imdb_movies['total_user_rating'].str.extract(r'(.*)[A-Z]')
imdb_movies['total_user_rating_0'] = imdb_movies['total_user_rating_0'].astype(float)
imdb_movies['total_user_rating_1'] = imdb_movies['total_user_rating'].str.extract(r'.*([A-Z])')
imdb_movies['total_user_rating_2'] = imdb_movies['total_user_rating_1'].apply(lambda x: 1000000 if x == 'M' else 1000)
imdb_movies['total_user_rating'] = imdb_movies['total_user_rating_0'] * imdb_movies['total_user_rating_2'] 
```

```python
imdb_movies
```

```python
imdb_movies['wins'] = imdb_movies['achievements'].str.extract(r'(\d*) wins')
imdb_movies['wins'] = imdb_movies['wins'].fillna(0)
imdb_movies['wins'] = imdb_movies['wins'].astype(int)
imdb_movies['nominations'] = imdb_movies['achievements'].str.extract(r'(\d*) nomination')
imdb_movies['nominations'] = imdb_movies['nominations'].fillna(0)
imdb_movies['nominations'] = imdb_movies['nominations'].astype(int)
imdb_movies['achievements'] = imdb_movies['wins']+imdb_movies['nominations']
imdb_movies
```

```python
imdb_movies = imdb_movies.loc[:,'title':'gross_worldwide']
imdb_movies
```

```python
imdb_movies.info()
```

```python
imdb_movies['budget']=imdb_movies['budget'].str.replace(',','')
imdb_movies['gross_worldwide']=imdb_movies['gross_worldwide'].str.replace(',','')
```

```python
imdb_movies
```

```python
imdb_movies['total_user_rating']=imdb_movies['total_user_rating'].astype(int)
imdb_movies['budget']=imdb_movies['budget'].astype(float)
imdb_movies['gross_worldwide']=imdb_movies['gross_worldwide'].astype(float)
imdb_movies.info()
```

```python
imdb_movies = imdb_movies.fillna('-')
```
